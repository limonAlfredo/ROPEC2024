## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Bearing Analysis

```python
dir_csv = './data/bearings'

dataframes = []

for file in os.listdir(dir_csv):
    if file.endswith('.csv'):
        path = os.path.join(dir_csv, file)
        df = pd.read_csv(path)
        dataframes.append(df)

bearings_df = pd.concat(dataframes, ignore_index=True)
bearings_df = bearings_df.dropna()
bearings_df.head()
```

## Motor Analysis

```python
dir_csv = './data/motor'

dataframes = []

for file in os.listdir(dir_csv):
    if file.endswith('.csv'):
        path = os.path.join(dir_csv, file)
        df = pd.read_csv(path)
        dataframes.append(df)

motor_df = pd.concat(dataframes, ignore_index=True)
motor_df = motor_df.dropna()
motor_df.head()
```

```python
mask = [True if isinstance(data,str) else False for data in motor_df.accx ]
motor_df[mask].head()
```

```python
motor_df = motor_df[motor_df != "None"]
```

```python
motor_df["accx"] = motor_df["accx"].astype(float)
motor_df["accy"] = motor_df["accy"].astype(float)
motor_df["accz"] = motor_df["accz"].astype(float)
motor_df["datetime"] = pd.to_datetime(motor_df["datetime"])
```
